# The Haar Wavelet Transform

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def fht(x, n):
    """
    Haar Wavelet Transform
    n = 2^t
    y = Wn * x
    """
    
    if n == 1:
        return x
    m = n // 2
    y = np.zeros((n, 1))
    z = fht(x[:m], m)
    y[::2] = z + x[m:]
    y[1::2] = z - x[m:]
    return y
```

```python
x = np.array([9, 7, 3, 5, 6, 10, 2, 6]).reshape((8, 1))
fht(x, 8)
```
